```python
import os
import re
import pandas as pd
```

```python
%reload_ext autoreload
%autoreload 2
import extraction_methods as em
```

```python
email_files = em.get_all_text_files('C:\\Users\\alexp\\Documents\\Програмач\\Python\\Email Classification Challenge\\email_examples_eplus\\NEUEBEISPIELMENGE')
```

```python
email_classes = [em.get_class_from_path(email_path) for email_path in email_files]
```

```python
email_texts = [em.extract_email_text(email_file) for email_file in email_files]
```

```python
email_texts
```

```python
tmp_dict = {'label': email_classes, 'text': email_texts}
df = pd.DataFrame(tmp_dict)
```

```python
df
```

```python
blanks = []
for i, record in enumerate(df['text']):
    if not record:
        blanks.append(i)
```

```python
df.drop(blanks, inplace=True)
```

```python
df
```

```python
category_counts = df['label'].value_counts()
threshold = 20
categories_to_delete = category_counts[category_counts < threshold].index.tolist()
df = df[~df['label'].isin(categories_to_delete)]
```

```python
from sklearn.model_selection import train_test_split
X = df['text']
y = df['label']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import LabelEncoder

tfidf_vectorizer = TfidfVectorizer()
X = tfidf_vectorizer.fit_transform(X)

label_encoder = LabelEncoder()
y = label_encoder.fit_transform(y)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)

rfc = RandomForestClassifier(n_estimators=100)

rfc.fit(X_train, y_train)
```

```python
rfc_pred = rfc.predict(X_test)
```

```python
print(classification_report(y_test, rfc_pred))
```
